```python
%matplotlib notebook
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
def fibonacci(n: int, gr=False, ls=False):
    nums = [1,1]
    for i in range(2,n):
        num = nums[i-1]+nums[i-2]
        nums.append(num)
    if gr:
        fig = plt.figure(figsize=(8,3) )
        ax1 = fig.add_subplot()
        ax1.plot(nums,color="r",label="Serie de Fibonacci")
        ax1.legend(loc=1)
    if ls:
        print(nums)
    return nums[n-1]
```

```python
fibonacci(20,gr=True,ls=True)
```

```python
def aureo(p: int):
    """Calcula el valor del nÃºmero Ã¡ureo con una exactitud proporcional a p.
    """
    au = fibonacci(p+1)/fibonacci(p)
    return au
```

```python
print(aureo(500))
```
