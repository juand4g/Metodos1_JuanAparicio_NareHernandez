```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
```

```python
def es_primo(num):
    for n in range(2, num):
        if num % n == 0:
            return False
    return True
```

```python
def primos(n: int):
    """Calcula n números primos.
    """
    primos = []
    
    i = 2
    while len(primos) < n:
        if es_primo(i):
            primos.append(i)
        i+=1
    
    return primos        
```

```python
lista = primos(1000)
print("Números de primos calculados: " + str(len(lista)))
print("\nPrimeros 10 primos calculados:")
for n in range(1,11):
    print(str(n) + ". " + str(lista[n-1]))
```

```python
fig = plt.figure(figsize=(8,3) )
ax1 = fig.add_subplot()
ax1.plot(lista,color="#572364",label="Primeros 1000 números primos")
ax1.legend(loc=1)
```
