```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
```

$$ r = a + b\theta $$

```python
a = 0
b = 1
```

```python
def arquimedes(n = 500):
    t = np.linspace(0,6*np.pi,n)
    equis = []
    ye = []
    
    for i in t:
        r = a + b*i
        x = r*np.cos(i)
        y = r*np.sin(i)
        equis.append(x)
        ye.append(y)
        
    fig = plt.figure(figsize=(8,3) )
    ax1 = fig.add_subplot()
    ax1.plot(equis,ye,color="b",label="Espiral de Arquímedes")
    ax1.legend(loc=1)
```

```python
arquimedes()
```
