```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import argrelextrema
```

```python
def maximos(route: str):
    archivo = open(route, "r")
    valoresX = []
    valoresY = []
    
    Xmax = []
    Ymax = []

    for linea in archivo:
        valores = linea.split()
        valX = float(valores[0])
        valY = float(valores[1])

        valoresX.append(valX)
        valoresY.append(valY)
    
    npY = np.array(valoresY)
    npX = np.array(valoresX)
    maxs = argrelextrema(npY, np.greater)

    for h in range(len(maxs[0])):
        i = maxs[0][h]
        Ymax.append(npY[i])
        Xmax.append(valoresX[i])
    plt.plot(valoresX,valoresY)
    plt.scatter(Xmax,Ymax,color="r",label="Máximos locales")
    plt.legend(loc=0)
    plt.show()
```

```python
ruta = "Data/EstrellaEspectro.txt"
maximos(ruta)
```
