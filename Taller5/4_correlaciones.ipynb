```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
```

```python
def correlacion(k,N,X):
    c = 1/N
    suma=0
    for i in range(N):
        if ((k+i)>=N):
            suma+=X[i]*X[i-k]
        else:
            suma+=X[i]*X[i+k]        
    c*=suma
    return c
```

```python
K=np.linspace(1,30,30)
N=10**4
X=np.random.uniform(size=(N,))
C=[]

for k in K:
    c = correlacion(int(k),N,X)
    C.append(c)
```

```python
plt.plot(K,C,color="r")
plt.ylim([0.20,0.30])
plt.title("Generador numpy",size=16)
plt.axhline(y=(1/4),color="#000000",label="Valor teórico",ls="--")
plt.legend()
plt.grid()
plt.xlabel("k-ésimo vecino", size = 16)
plt.ylabel("C(k)", size = 16)
```
