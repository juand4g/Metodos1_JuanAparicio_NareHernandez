```python
import matplotlib.pyplot as plt
import sympy as sym
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
```

```python
p = np.array([[1,2],[3,4]])
xs = sym.Symbol('x',Real=True)
ys = sym.Symbol('y',Real=True)
    
polym = 0
for i in range(2):
    for j in range(2):
        polym += p[i,j]*(xs**i)*(ys**j)
polym
```

```python
def Temperatura(position, P):
    x = position[0]
    y = position[1]
    
    T = P[0] + P[1]*x + P[2]*y + P[3]*x*y
            
    return T
```

```python
def Solve(A,B):
    P = np.dot((np.linalg.inv(A)),B)
    
    return P
```

```python
position = np.zeros((4,2)) 
position[0,0]=1
position[0,1]=1

position[1,0]=-1
position[1,1]=1

position[2,0]=-1
position[2,1]=-1

position[3,0]=1
position[3,1]=-1
```

```python
for i in range(4):
    for j in range(2):
        position

position
```

```python
A = np.zeros((4,4))
A[:,0] = 1
A[:,1] = position[:,0]
A[:,2] = position[:,1]
A[:,3] = position[:,0]*position[:,1]
B = np.array([[1],[2],[0.5],[0.3]])
A
```

```python
B
```

```python
P=Solve(A,B)
P
```

```python
for i in range(4):
    print(Temperatura(position[i],P))
```

```python
T = lambda x,y: P[0] + P[1]*x + P[2]*y + P[3]*x*y
```

```python
x=np.linspace(-1,1,100)
y=np.linspace(-1,1,100)
X,Y = np.meshgrid(x,y)
Z = T(X,Y)
fig = plt.figure(figsize=(8,3))
ax = fig.add_subplot(1,2,1,projection='3d')
ax1 = fig.add_subplot(1,2,2)

ax.plot_surface(X,Y,Z)
ax1.contour(X,Y,Z,100)
```

```python
def Rotate(Positions,theta):
    rotated = np.zeros((4,4))
    
    for i in range(4):
        pos = Positions[i]
        pos_ = np.zeros((2,1))
        pos_[0]=pos[0]
        pos_[1]=pos[1]
        m = np.array([[np.cos(theta),-np.sin(theta)],[np.sin(theta),np.cos(theta)]])
        rpos = np.dot(m,pos_)
        
        rotated[0,0] = rpos[0]
        rotated[0,1] = rpos[1]
        
    return rotated
Rotate(position, 1.03)
```

```python
theta = np.linspace(0,2*np.pi,200)
temperaturas = np.array((200))

for i in range(200):
    t = theta[i]
    rpos = Rotate(position,t)
    rA = np.zeros((4,4))
    rA[:,0] = 1
    rA[:,1] = rpos[:,0]
    rA[:,2] = rpos[:,1]
    rA[:,3] = rpos[:,0]*rpos[:,1]
    print(rA)
    rp=Solve(rA,B)
    rT = lambda x,y: rp[0] + rp[1]*x + rp[2]*y + rp[3]*x*y
    k = rT(0,0.5)
    temperaturas[i] = k
    
minim = np.amin(temperaturas)
minim
    
```

```python
def Rotar(Posicion,angulo):
    matrizrotadora= [[np.cos(angulo), -np.sin(angulo)],[np.sin(angulo), np.cos(angulo)]]
    posicionrotadav0 = Posicion[0]*matrizrotadora
    posicionrotadav1 = Posicion[1]*matrizrotadora
    posicionrotadav2 = Posicion[2]*matrizrotadora
    posicionrotadav3 = Posicion[3]*matrizrotadora
    Posicion[1]=  posicionrotadav1 
    Posicion[0]=  posicionrotadav0    
    Posicion[2]=  posicionrotadav2
    Posicion[3]=  posicionrotadav3
    return Posicion
    
```
