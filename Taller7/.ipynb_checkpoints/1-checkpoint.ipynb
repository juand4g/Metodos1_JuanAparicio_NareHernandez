```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
```

$$2x-y=2$$
$$x+2y=1$$
$$x+y=4$$

```python
l1 = lambda x: 2*x-2
l2 = lambda x: (1-x)/2
l3 = lambda x: 4-x
```

## MÍNIMOS CUADRADOS

```python
def minCuad(A,B):
    m,n = A.shape
   
    AtA = np.dot(A.T,A)
    
    AtB = np.dot(A.T,B)
    
    X = np.linalg.solve(AtA,AtB)
    
    D = np.dot(A,X)-B
    
    return X,D
```

```python
A = np.array([[2,-1],[1,2],[1,1]])
B = np.array([[2],[1],[4]])
xy,d = minCuad(A,B)
xy
```

```python
d
```

```python
np.linalg.norm(d)
```

```python
dom = np.linspace(-5,5,200)

plt.plot(dom,l1(dom),label="Línea 1")
plt.plot(dom,l2(dom),label="Línea 2")
plt.plot(dom,l3(dom),label="Línea 3")
plt.scatter(xy[0],xy[1],label="Solución",c="r")
plt.legend()
```

## RESUMEN MIN. CUADRADOS
Se obtuvieron los valores
$$x=1.42857143$$
$$y=0.42857143$$

como la mejor solución aproximada al sistema de ecuaciones planteado. Con dichos valores, se obtuvo una distancia de

$$d=2.5354627641855494$$

## BÚSQUEDA ITERATIVA

```python
x = np.linspace(-5,5,333)
y = np.linspace(-5,5,333)

D = np.zeros((333,333))

mind = 99999999999

for i in range(x.size):
    for j in range(y.size):
        TMP = np.array([[x[i]],[y[j]]])
        d = np.dot(A,TMP)-B
        dist = np.linalg.norm(d)
        D[i,j] = dist
        
X,Y = np.meshgrid(x,y)

fig = plt.figure(figsize=(16,9))
ax = fig.add_subplot(projection='3d')
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("Distancia")
surf = ax.plot_surface(X, Y, D, cmap=cm.coolwarm,
                       linewidth=0, antialiased=True)   
fig.colorbar(surf, shrink=0.5, aspect=5)
ax.view_init(30, 45)
```

## RESUMEN GENERAL
Los puntos obtenidos por mínimos cuadrados y por búsqueda iterativa son básicamente los mismos. La diferencia entre los puntos obtenidos en cada método fue de:
$$|x_{min. cuadrados}-x_{búsqueda}|=|1.42857143-1.4156626506024095|=0.0129...$$
$$|y_{min. cuadrados}-y_{búsqueda}|=|0.42857143-0.4216867469879517|=0.00688...$$

y la distancia obtenida en ambos casos, por lo tanto, también es aproximadamente la misma:

$$2.535751049799306\approx2.5354627641855494$$

De este modo, es posible evidenciar que el resultado obtenido con el método de mínimos cuadrados es, efectivamente, la solución que minimiza la distancia.
