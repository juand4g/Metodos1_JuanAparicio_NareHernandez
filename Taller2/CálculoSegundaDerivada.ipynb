```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
```

```python
def SegundaDerivadaCentral(x,f,h):
    SecDerivada= (f(x+h)-(2*f(x))+ f(x-h))/(h**2)
    return SecDerivada
```

```python

x= np.linspace(-10.0,10.0,400)
expmenosxalados= lambda x: ((1+np.exp(-x**2))**-0.5)
print(SegundaDerivadaCentral(x,expmenosxalados,0.05))
```
