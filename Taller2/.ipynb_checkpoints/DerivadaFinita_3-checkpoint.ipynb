```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
import sympy as sym
```

```python
def ConvolutionDerivative(ran,h,m=[-1,0,1]):
    dydx = [0]
    for n in range(1,ran.size-1):
        k = m[0]*ran[n-1] + m[1]*ran[n] + m[2]*ran[n+1]
        k = k*(1/2*h)
        dydx.append(k)
    dydx.append(0)
    
    DyDx = np.array(dydx)
    
    return DyDx 
```

```python
x = sym.Symbol("x",Real=True)
function = 1/sym.sqrt(1+sym.exp(-x**2))
f = sym.lambdify([x],function,"numpy")

canonDerivative = (sym.exp(-x**2)*x)/(1+sym.exp(-x**2))**(3/2)
canonDyDx = sym.lambdify([x],canonDerivative,"numpy")
```

```python
function
```

```python
dom = np.linspace(-10,10,100)
ran = f(dom)
DyDx = ConvolutionDerivative(ran,h=5)
canDyDx = canonDyDx(dom)
```

```python
fig = plt.figure(figsize=(8,8))

ax1 = fig.add_subplot(3,1,1)
ax1.plot(dom,ran,color="b",label="f(x)")
ax1.legend(loc=3)

ax2 = fig.add_subplot(3,1,2)
ax2.plot(dom,DyDx,color="r",label="f'(x)")
ax2.plot(dom,canDyDx,color="#F44611",label="f'(x) canon")
ax2.legend(loc=3)

ax3 = fig.add_subplot(3,1,3)
ax3.scatter(dom,np.abs(DyDx-canDyDx),c="#000000",label="Error")
ax3.legend(loc=1)
```
